Aca voy a realizar el ajuste de las curvas al inicio de la epidemia y voy a calcular el exponente de crecimiento que se relaciona con el $R_0$

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)
import seaborn as sns
```

Cargo los valores de los resultados para distintos valores de $P$, tengo susceptibles, infectados y recuperados para cada probabilidad.

```python
results=np.load('results_nuevo_600.npy')
```

```python
plt.figure(2)
for i in range(len(results)):
  plt.plot(range(1,250+1), results[i][1], label=f'Probabilidad de infeccion={(i+1)/10}')
plt.xlabel('Días')
plt.ylabel('Número de personas infectadas (log scale)')
plt.title('Dinamica de la epidemia para distintos valores P')
#plt.yscale("log")
#plt.ylim(0,10**7)
plt.legend(fontsize = 'small')
plt.show()
```

Voy a usar sklearn para ajustar a rectas el comienzo del crecimiento de las curvas luego del tiempo transitorio en el cual el tiempo se acomoda. Primero observo que para los valores de $P$ menores a $0.4$ no existe crecimiento exponencial por lo cual no voy a poder calcular el $R_0$ con este metodo.

```python
#Miro uno por uno las curvas para determinar cuando termina el tiempo transitorio
#P=1
plt.figure(3)
for p in range(0,3):
    plt.plot(range(1,250+1)[:250], results[p][1][:250], label=f'Probabilidad de infeccion={(p+1)/10}')
plt.xlabel('Días')
plt.ylabel('Número de personas infectadas (log scale)')
plt.title('Dinamica de la epidemia para distintos valores P')
#plt.yscale("log")
plt.legend()
plt.show()
```

Al ajustar a una recta debo recordar que la pendiente de la recta $\lambda$, quien a su vez es el exponente de crecimiento al comienzo de la epidemia se relaciona con el $R_0$ como $\lambda= \frac{R_0 -1}{\tau}$ entonces obtenemos el numero promedio de infecciones secundarias como $R_0 = \lambda \tau + 1$, aquí utilizamos $\tau = 10$

```python
plt.figure(3)
plt.plot(range(1,250+1)[:30], results[3][1][:30], label=f'Probabilidad de infeccion=0.4')
plt.xlabel('Días')
plt.ylabel('Número de personas infectadas (log scale)')
plt.title('Dinamica de la epidemia para distintos valores P')
plt.yscale("log")
plt.legend()
plt.show()
```

```python
#Voy a ajustar una recta entre el dia 5 y el dia 20
R_0_exponente=[]
tau=10
from sklearn.linear_model import LinearRegression
x=np.arange(18,30).reshape(-1,1)
y=np.log(results[3][1][18:30])
model=LinearRegression().fit(x,y)
R_0_exponente.append(model.coef_[0]*tau+1)
```
